```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import numpy as np
import nltk
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
def data_processing(data):
    #Lowercase conversion
    data = data.lower()
    #tokenize the data
    data_tokens = word_tokenize(data)
    #remove_stopwords
    processed_words = [w for w in data_tokens if not w in stop_words]
    return" ".join(processed_words)
```

```python
data2 = pd.read_csv('Youtube04 Eminem.csv')
print(data2)
```

```python
data2.CONTENT = data2['CONTENT'].apply(data_processing)
```

```python
content_text = ' '.join([w for w in data2['CONTENT']])
plt.figure(figsize =(10,10), facecolor='None')
wordcloud = WordCloud().generate(content_text)
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
