```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
```

```python
flights_data = sns.load_dataset('flights')
```

```python
flights_data.head()
```

```python
flights_data.describe()
```

```python
flights = flights_data.pivot("year", "month", "passengers")
plt.figure(figsize=(10,8))
ax = sns.heatmap(flights, cbar_kws={"orientation":"horizontal"})
plt.title('Heatmap of flight data')
plt.show()
```

```python
flights = flights_data.pivot("year", "month", "passengers")
plt.figure(figsize=(10,8))
ax = sns.heatmap(flights, linewidths=0.5, annot=True, fmt="d")
plt.title('Heatmap of flight data')
plt.show()
```

```python
flights = flights_data.pivot("year", "month", "passengers")
plt.figure(figsize=(10,8))
ax = sns.heatmap(flights, linewidths=0.5,linecolor="black", annot=True, fmt="d")
plt.title('Heatmap of flight data')
plt.show()
```

```python
flights = flights_data.pivot("year", "month", "passengers")
plt.figure(figsize=(10,8))
ax = sns.heatmap(flights, linewidths=0.5,linecolor="black", annot=True, fmt="d", cmap="coolwarm")
plt.title('Heatmap of flight data')
plt.show()
```

```python
flights = flights_data.pivot("year", "month", "passengers")
plt.figure(figsize=(10,8))
ax = sns.heatmap(flights, linewidths=0.5,linecolor="black", annot=True, fmt="d", cmap="magma")
plt.title('Heatmap of flight data')
plt.show()
```

```python
#Correlation Method using Heatmap

from sklearn.datasets import fetch_california_housing
housing_data = fetch_california_housing()
```

```python
df = pd.DataFrame(housing_data.data)
df.columns = housing_data.feature_names
df['Pricee'] = housing_data.target
df.head()
```

```python
print(housing_data.DESCR)
```

```python
df.corr()
```

```python
plt.figure(figsize=(10,7))
ax = sns.heatmap(df.corr(), annot=True)
plt.title('California housing data')
```

```python
#Only The Lower Correlation Matrix
plt.figure(figsize=(10,7))
mask = np.triu(np.ones_like(df.corr()))
ax = sns.heatmap(df.corr(), annot=True, mask = mask, annot_kws={'fontsize':12, 'fontweight':'bold'}, linewidths=0.5, fmt='0.3f')
plt.title('California housing data')
```
